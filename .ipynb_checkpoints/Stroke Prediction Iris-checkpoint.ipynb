<div style="margin-top:10px;" id=domain-understanding><h2>Domain Understanding</h2></div>
<p style="text-align: justify;">To be able to predict whether someone will have a stroke, we first need to do some research about stroke in general. </p><br>

<b> What is a stroke?</b>
<p style="text-align: justify; margin-top:0;">A stroke (or brain attack) happens when either something blocks the blood supply to the brain or when a blood vessel in the brain bursts. </p><br>

1. <u>Ischemic stroke:</u>
<p style="text-align: justify; margin-top:0;">This happens when blood clots or other particles block the blood vessels to the brain. This type of stroke is the most common one.</p><br>

2. <u>Hemorrhagic stroke:</u>
<p style="text-align: justify; margin-top:0;">This stroke happens when an artery in the brain leaks blood or breaks open.</p><br>

3. <u>Transient ischemic stroke (TIA):</u>
<p style="text-align: justify; margin-top:0;margin-right: 100px;">Also called a 'mini-stroke'. When you have a TIA, it is basically a warning sign of a future stroke. The symptoms of a major stroke are similar to the symtoms of a TIA. This type of stroke is often caused by blood clots (like for ischemic stroke). Even though this type of stroke only occurs for a few minutes, it is very important to treat it, as more than third of people who had a TIA got a major stroke within 1 year. For 10% to 15% this was even within 3 months.</p>

<p style="text-align:justify;margin-right:100px;margin-top:20px;">
<b>Symptoms</b><br>
The most common symptoms are:
<ul>
    <li>Sudden numbness or weakness in the face, arm, or leg, especially on one side of the body.</li>
    <li>Sudden confusion, trouble speaking, or difficulty understanding speech.</li>
    <li>Sudden trouble seeing in one or both eyes.</li>
    <li>Sudden trouble walking, dizziness, loss of balance, or lack of coordination.</li>
    <li>Sudden severe headache with no known cause.</li>
</ul>
</p>

<p style="text-align:justify;margin-right:100px;margin-top:20px;">
<b>What could increase the chance of having a stroke?</b><br>
There are some things that can increase the chance of having a stroke. 
To give a few examples:
<ul>
    <li style="text-align:justify;margin-right:100px;">Hypertension: high blood pressure damages arteries throughout the body, creating conditions in which they can burst or clog more easily. Weakened or blocked arteries in the brain put you at a much higher risk for stroke, which is why managing high blood pressure is critical to reducing your stroke risk.</li>
    <li style="text-align:justify;margin-right:100px;">Heart disease: common heart disorders can increase your risk for stroke. For example, coronary artery disease increases your risk for stroke, because plaque builds up in the arteries and blocks the flow of oxygen-rich blood to the brain. Other heart conditions, such as heart valve defects, irregular heartbeat (including atrial fibrillation), and enlarged heart chambers, can cause blood clots that may break loose and cause a stroke.</li>
    <li style="text-align:justify;margin-right:100px;">Smoking: damages blood vessels, which can cause a stroke. About 1 in 7 men smoke. Men are also more likely to be smokers than women are.</li>
    <li style="text-align:justify;margin-right:100px;">Overweight and obesity: having overweight or obesity increases stroke risk. About 3 in 4 men in the United States have overweight or obesity.</li>
    <li style="text-align:justify;margin-right:100px;">Diabetes: increases stroke risk because it can harm blood vessels in the brain. About 1 in 7 men have diabetes.</li>
    <li style="text-align:justify;margin-right:100px;">Age: you are more likely to have a stroke if you are older. It is even so that the chance of having a stroke doubles every 10 years after age 55. However, nowadays it is so that one in seven strokes occur in adolescents and young adults ages 15-49, this could be because more young people have obesity, high blood pressure and diabetes.</li>
    <li style="text-align:justify;margin-right:100px;">Sex: women are more likely to have a stroke, this is because pregnancy and birth control pills have an extra risk.</li>
</ul>
</p>
<p style="text-align:justify;margin-right:100px;margin-top:20px;">
<b>What to do when you notice any of the symptoms?</b><br>
If a person might have a stroke, think <b>F.A.S.T.</b>:
<ul>
    <li><u><b>F</b>ace:</u> ask the person to smile to check whether one side of the face is drooping.</li>
    <li><u><b>A</b>rms:</u> ask the person to raise both arms, and see if one arm drifts downwards.</li>
    <li><u><b>S</b>peech:</u> ask the person to repeat a short sentence to find out if their words sounds slurred.</li>
    <li><u><b>T</b>ime:</u> if you notice any of these signs, call the emergency assistance immediately.<br><br></li>
</ul>

<b>Treatment</b><br>
<p style="text-align:justify;margin-right: 100px;">It is crucial to call the emergency assistence for help immediately, as every minute counts. The faster a stroke is treated, the less damage is done to the brain. Some treatments can only be used if the person reaches the hospital in the first three hours after symptoms appear.</p>

<div style="margin-top:20px;margin-bottom:-20px;" id=analytic-approach><h2>Analytic Approach</h2></div><br>
<p style="text-align:justify;margin-right: 100px;">Since we want to predict whether a patient is likely to have a stroke based on certain symptoms, we should look at the column <b>stroke</b>. This column indicates if the patient in the dataset eventually had a stroke or not. So stroke is our target variable. </p>

<p style="text-align:justify;margin-right: 100px;">    
Since we want to predict how big the risk is someone will have a stroke, we will use regression. I will start by using the algorithm KNN, as this was recommended.
</p>

<p style="text-align:justify;margin-right: 100px;">    
The target variable is binary, someone either had a stroke (1) or did not (0). A good method to use is logistic regression. For my challenge I will use this. Logistic regression is used to predict the output of a categorical dependent variable. The output therefore must be categorical or discrete. Even though it is called logistic regression, logistic regression is used for classification problems (as it is used to classify samples) and not for regression problems. For logistic regression it is so that either one of two max values is predicted (0 or 1). If the predicted value is 0.499, the outcome will be 0 and therefore predicts that the patient did not have a stroke. Instead of a straight line (for linear regression), we get a "S" shaped function.
</p>

```python
df.loc[df['gender'] == 'Female', 'gender'] = 1
df.loc[df['gender'] == 'Male', 'gender'] = 0
```

The column ever_married, work_type, Residence_type and smoking_status are probably not that important, but we will still keep these for now. We will change the datatypes, by using one-hot encoding.

```python
df = pd.get_dummies(df, columns = ['ever_married', 'work_type', 'Residence_type', 'smoking_status'])
df
```

When we will try to predict something, the machine will almost always predict 'no', since we have much more of those in our datasets. Our dataset is really imbalanced now and we should make it more balanced. We will downsample our data, so we will remove a lot of rows of patients who did not have a stroke.</p>

```python
from sklearn.utils import resample

df_majority = df[df['stroke']==0]
df_minority = df[df['stroke']==1]
 
# Downsample majority class
df_majority_downsampled = resample(df_majority, 
                                 replace=False,    # sample without replacement
                                 n_samples=209,    # to match minority class
                                 random_state=0)   # to keep the same values everytime we run the notebook
                                 
df_downsampled = pd.concat([df_majority_downsampled, df_minority])
 
df_downsampled.stroke.value_counts()
```

<h3>Logistic Regression</h3><p style="text-align:justify;margin-right:100px;">We now continue with logistic regression as this algorithm is good for predicting yes or no outcomes. We start by selecting features we decided are good.</p>

```python
X = df_downsampled[['age', 'ever_married_Yes', 'hypertension', 'avg_glucose_level', 'work_type_children']] 

y = df_downsampled["stroke"]
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)
```

```python
model1 = LogisticRegression(solver='liblinear', random_state=0)
```

```python
model1.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
pred = model1.predict(X_test)
report = classification_report(y_test, pred)
print(report)
```

<p style="text-align:justify;margin-right:100px;">The selected features gave the best results. Before we actually apply Logistic Regression, we want to know what parameters will give us the best results. To do this, we perform grid search.</p>

```python
pipelineLR = make_pipeline(StandardScaler(), LogisticRegression(random_state=1, penalty='l2', solver='lbfgs')) 

# Create the parameter grid
param_grid_lr = [{
    'logisticregression__C': [0.03, 0.04, 0.05, 0.1, 0.5, 1],
}]

# Create an instance of GridSearch Cross-validation estimator
gsLR = GridSearchCV(estimator=pipelineLR, 
                     param_grid = param_grid_lr,
                     scoring='average_precision', 
                     cv=10,
                     refit=True,
                     n_jobs=1)

# Train the LogisticRegression Classifier
gsLR = gsLR.fit(X_train, y_train)

# Print the training score of the best model
print(gsLR.best_score_)

# Print the model parameters of the best model
print(gsLR.best_params_)

# Print the test score of the best model
clfLR = gsLR.best_estimator_
print('Test accuracy: %.3f' % clfLR.score(X_test, y_test))
```

<p style="text-align:justify;margin-right:100px;">For the GridSearch we used a pipeline, since we want to use StandardScaler and LogisticRegression for the estimator. We can only apply one value to the estimator, so we add those in a pipeline. Since we want to improve the precision of the model, we use 'average_precision' for scoring.</p>

```python
model = LogisticRegression(solver='liblinear', random_state=0, C=0.04)
```

```python
model.fit(X_train, y_train)
```

<p style="text-align:justify;margin-right:100px;">According to the grid search, the best model needs a hyperparameter (C) of 0.05 (by default our hyperparameter is 1.00). We also apply solver='liblinear', which is good to use for small datasets.</p> <a href="https://ibmsoe.github.io/snap-ml-doc/sklogregdoc.html">https://ibmsoe.github.io/snap-ml-doc/sklogregdoc.html</a>

```python
# probability estimates
model.predict_proba(X)
```

<p style="text-align:justify;margin-right:100px;">Now that we chose our hyperparameter, we can see that the model is guessing a lot, as many of them are close to 0.5. (Left is no, right is yes) It is mostly guessing 1 (stroke), which is better than predicting no stroke, when someone did have a stroke. Everything that was close to 1 when making the prediction, will be put to 1. But everything that is 0.5 or lower, will be predicted as no. If the result is around 0.55 or 0.6, it tried really hard to changed the value to no, but this did not work, which means if was a clear yes.</p>

```python
model.predict(X)
```

<p style="text-align:justify;margin-right:100px;">Above we can see what actually was predicted to be the outcome. We see that most of them are 1 (stroke).</p>

```python
model.score(X, y)
```

Because we chose a hyperparameter and tried to get the best results possible, either recall or precision will get better results. Since one of them gets better, the other one will get worse.

<div style="margin-top:10px;" id=domain-understanding><h2>Domain Understanding</h2></div>
<p style="text-align: justify;">To be able to predict whether someone will have a stroke, we first need to do some research about stroke in general. </p><br>

<b> What is a stroke?</b>
<p style="text-align: justify; margin-top:0;">A stroke (or brain attack) happens when either something blocks the blood supply to the brain or when a blood vessel in the brain bursts. </p><br>

1. <u>Ischemic stroke:</u>
<p style="text-align: justify; margin-top:0;">This happens when blood clots or other particles block the blood vessels to the brain. This type of stroke is the most common one.</p><br>

2. <u>Hemorrhagic stroke:</u>
<p style="text-align: justify; margin-top:0;">This stroke happens when an artery in the brain leaks blood or breaks open.</p><br>

3. <u>Transient ischemic stroke (TIA):</u>
<p style="text-align: justify; margin-top:0;margin-right: 100px;">Also called a 'mini-stroke'. When you have a TIA, it is basically a warning sign of a future stroke. The symptoms of a major stroke are similar to the symtoms of a TIA. This type of stroke is often caused by blood clots (like for ischemic stroke). Even though this type of stroke only occurs for a few minutes, it is very important to treat it, as more than third of people who had a TIA got a major stroke within 1 year. For 10% to 15% this was even within 3 months.</p>

<p style="text-align:justify;margin-right:100px;margin-top:20px;">
<b>Symptoms</b><br>
The most common symptoms are:
<ul>
    <li>Sudden numbness or weakness in the face, arm, or leg, especially on one side of the body.</li>
    <li>Sudden confusion, trouble speaking, or difficulty understanding speech.</li>
    <li>Sudden trouble seeing in one or both eyes.</li>
    <li>Sudden trouble walking, dizziness, loss of balance, or lack of coordination.</li>
    <li>Sudden severe headache with no known cause.</li>
</ul>
</p>

<p style="text-align:justify;margin-right:100px;margin-top:20px;">
<b>What could increase the chance of having a stroke?</b><br>
There are some things that can increase the chance of having a stroke. 
To give a few examples:
<ul>
    <li style="text-align:justify;margin-right:100px;">Hypertension: high blood pressure damages arteries throughout the body, creating conditions in which they can burst or clog more easily. Weakened or blocked arteries in the brain put you at a much higher risk for stroke, which is why managing high blood pressure is critical to reducing your stroke risk.</li>
    <li style="text-align:justify;margin-right:100px;">Heart disease: common heart disorders can increase your risk for stroke. For example, coronary artery disease increases your risk for stroke, because plaque builds up in the arteries and blocks the flow of oxygen-rich blood to the brain. Other heart conditions, such as heart valve defects, irregular heartbeat (including atrial fibrillation), and enlarged heart chambers, can cause blood clots that may break loose and cause a stroke.</li>
    <li style="text-align:justify;margin-right:100px;">Smoking: damages blood vessels, which can cause a stroke. About 1 in 7 men smoke. Men are also more likely to be smokers than women are.</li>
    <li style="text-align:justify;margin-right:100px;">Overweight and obesity: having overweight or obesity increases stroke risk. About 3 in 4 men in the United States have overweight or obesity.</li>
    <li style="text-align:justify;margin-right:100px;">Diabetes: increases stroke risk because it can harm blood vessels in the brain. About 1 in 7 men have diabetes.</li>
    <li style="text-align:justify;margin-right:100px;">Age: you are more likely to have a stroke if you are older. It is even so that the chance of having a stroke doubles every 10 years after age 55. However, nowadays it is so that one in seven strokes occur in adolescents and young adults ages 15-49, this could be because more young people have obesity, high blood pressure and diabetes.</li>
    <li style="text-align:justify;margin-right:100px;">Sex: women are more likely to have a stroke, this is because pregnancy and birth control pills have an extra risk.</li>
</ul>
</p>
<p style="text-align:justify;margin-right:100px;margin-top:20px;">
<b>What to do when you notice any of the symptoms?</b><br>
If a person might have a stroke, think <b>F.A.S.T.</b>:
<ul>
    <li><u><b>F</b>ace:</u> ask the person to smile to check whether one side of the face is drooping.</li>
    <li><u><b>A</b>rms:</u> ask the person to raise both arms, and see if one arm drifts downwards.</li>
    <li><u><b>S</b>peech:</u> ask the person to repeat a short sentence to find out if their words sounds slurred.</li>
    <li><u><b>T</b>ime:</u> if you notice any of these signs, call the emergency assistance immediately.<br><br></li>
</ul>

<b>Treatment</b><br>
<p style="text-align:justify;margin-right: 100px;">It is crucial to call the emergency assistence for help immediately, as every minute counts. The faster a stroke is treated, the less damage is done to the brain. Some treatments can only be used if the person reaches the hospital in the first three hours after symptoms appear.</p>

<div id=data-preparation><h2>Data Preparation</h2></div>
<p style="text-align:justify;margin-right: 50px;">
    Here we will change what we found during <a href="#data-understanding">Data Understanding</a>.
</p>

We want to have numerical data so we change Female to 0 and Male to 1 (alphabetical is first).

```python
df.loc[df['gender'] == 'Female', 'gender'] = 1
df.loc[df['gender'] == 'Male', 'gender'] = 0
```

The column ever_married, work_type, Residence_type and smoking_status are probably not that important, but we will still keep these for now. We will change the datatypes, by using one-hot encoding.

```python
df = pd.get_dummies(df, columns = ['ever_married', 'work_type', 'Residence_type', 'smoking_status'])
df
```

<p style="text-align:justify;margin-right:100px;">We now have 22 columns instead of the 12 that we started with. When we used one-hot encoding, the datatypes of the new columns changed to a datatype called "uint8", so this will be changed to int64. Also the column age_group will be changed to category.</p>

```python

df[["gender", "ever_married_No", "ever_married_Yes", "work_type_Govt_job", "work_type_Never_worked", 
    "work_type_Private", "work_type_Self-employed", "work_type_children", "Residence_type_Rural", 
    "Residence_type_Urban", "smoking_status_Unknown", "smoking_status_formerly smoked", 
    "smoking_status_never smoked", "smoking_status_smokes"]] = df[["gender", "ever_married_No", "ever_married_Yes", 
    "work_type_Govt_job", "work_type_Never_worked", "work_type_Private", "work_type_Self-employed", 
    "work_type_children", "Residence_type_Rural", "Residence_type_Urban", "smoking_status_Unknown", 
    "smoking_status_formerly smoked", "smoking_status_never smoked", "smoking_status_smokes"]].astype('int64')
df.info()
```

<p style="text-align:justify;margin-right:100px;">We will now create a new column called Age group so we can fill the missing values and make visualizations per group. Probably it is better to drop the data with age under 18, but we will look at that later.</p>

```python
df.loc[df['age'] < 18, 'age_group'] = '0-18'
df.loc[(df['age'] >= 18) & (df['age'] < 30), 'age_group'] = '18-30'
df.loc[(df['age'] >= 30) & (df['age'] < 50), 'age_group'] = '30-50'
df.loc[(df['age'] >= 50) & (df['age'] < 70), 'age_group'] = '50-70'
df.loc[df['age'] >= 70, 'age_group'] = '70+'
df['age_group'] = df['age_group'].astype('category')
df
```

<p style="text-align:justify;margin-right:100px;">The next step is to fill the missing values with the mean value for each gender in the grouped ages.</p>

```python
female_younger_than_18 = df.loc[(df['gender']==0) & (df['age_group']=='0-18')] # 438
male_younger_than_18 = df.loc[(df['gender']==1) & (df['age_group']=='0-18')] # 418 

female_between_18_30 = df.loc[(df['gender']==0) & (df['age_group']=='18-30')] # 222
male_between_18_30 = df.loc[(df['gender']==1) & (df['age_group']=='18-30')] # 436 

female_between_30_50 = df.loc[(df['gender']==0) & (df['age_group']=='30-50')] # 518
male_between_30_50 = df.loc[(df['gender']==1) & (df['age_group']=='30-50')] # 867

female_between_50_70 = df.loc[(df['gender']==0) & (df['age_group']=='50-70')] # 631
male_between_50_70 = df.loc[(df['gender']==1) & (df['age_group']=='50-70')] # 824

female_older_than_70 = df.loc[(df['gender']==0) & (df['age_group']=='70+')] # 306 
male_older_than_70 = df.loc[(df['gender']==1) & (df['age_group']=='70+')] # 449
male_between_18_30
```

<p style="text-align:justify;margin-right:100px;">We check the mean values of bmi of each of the new created dataframes. We also check how many missing values the column bmi in these dataframes has, now we can also check if it is true that we have 201 missing values.</p>

```python
print("Mean bmi F < 18: ", female_younger_than_18['bmi'].mean(), "- missing values:", female_younger_than_18['bmi'].isnull().sum())
print("Mean bmi M < 18: ", male_younger_than_18['bmi'].mean(), "- missing values:", male_younger_than_18['bmi'].isnull().sum())
print()
print("Mean bmi F 18-30:", female_between_18_30['bmi'].mean(), "- missing values:", female_between_18_30['bmi'].isnull().sum())
print("Mean bmi M 18-30:", male_between_18_30['bmi'].mean(), "- missing values:", male_between_18_30['bmi'].isnull().sum())
print()
print("Mean bmi F 30-50:", female_between_30_50['bmi'].mean(), " - missing values:", female_between_30_50['bmi'].isnull().sum())
print("Mean bmi M 30-50:", male_between_30_50['bmi'].mean(), "- missing values:", male_between_30_50['bmi'].isnull().sum())
print()
print("Mean bmi F 50-70:", female_between_50_70['bmi'].mean(), "- missing values:", female_between_50_70['bmi'].isnull().sum())
print("Mean bmi M 50-70:", male_between_50_70['bmi'].mean(), " - missing values:", male_between_50_70['bmi'].isnull().sum())
print()
print("Mean bmi F > 70: ", female_older_than_70['bmi'].mean(), " - missing values:", female_older_than_70['bmi'].isnull().sum())
print("Mean bmi M > 70: ", male_older_than_70['bmi'].mean(), "- missing values:", male_older_than_70['bmi'].isnull().sum())
```

<p style="text-align:justify;margin-right:100px;">For now we will keep the age group of 0-18, since it might be useful to understand our data better. However, the mean bmi value of female and male is around 21, which does not make sense for a baby or young child. For now we will fill the missing values with the mean value of the groups created above, rounded to one decimal. The reason we will takes out those rows, is because it is very hard to determine what is a good limit. The limit of people under 18 is different per gender and each age (the bmi of a 3 year old boy should be between 14 and 18, while the bmi of a 18 year old boy should be between 17 and 25). For children under 2 years old, bmi is not even used. So, to sum up, these rows should be dropped.</p>

```python
female_younger_than_18['bmi'] = female_younger_than_18['bmi'].fillna(21.3)
male_younger_than_18['bmi'] = male_younger_than_18['bmi'].fillna(21.5)
female_between_18_30['bmi'] = female_between_18_30['bmi'].fillna(27.8)
male_between_18_30['bmi'] = male_between_18_30['bmi'].fillna(28.0)
female_between_30_50['bmi'] = female_between_30_50['bmi'].fillna(31.7)
male_between_30_50['bmi'] = male_between_30_50['bmi'].fillna(31.1)
female_between_50_70['bmi'] = female_between_50_70['bmi'].fillna(31.6)
male_between_50_70['bmi'] = male_between_50_70['bmi'].fillna(31.3)
female_older_than_70['bmi'] = female_older_than_70['bmi'].fillna(28.9)
male_older_than_70['bmi'] = male_older_than_70['bmi'].fillna(29.2)
```

<p style="text-align:justify;margin-right:100px;">Since we filled the missing values in each dataframe seperately, we will merge them all together again. 
We now have the full dataset again, sorted by age group. We do not have any more missing values.</p>

```python
df = pd.concat(
    [female_younger_than_18, male_younger_than_18, 
     female_between_18_30, male_between_18_30, 
     female_between_30_50, male_between_30_50, 
     female_between_50_70, male_between_50_70, 
     female_older_than_70, male_older_than_70])
df.isnull().sum()
```

```python
print('The mean value of bmi: ', df['bmi'].mean())
```

<p style="text-align:justify;margin-right:100px;">The mean bmi is still around the same as before.</p>

```python
df['age'].hist(by=df['stroke'], bins=10, color='lightblue', edgecolor='black', linewidth='0.5', figsize=(12,8))
```

<p style="text-align:justify;margin-right:100px;">We can clearly that from a certain age the risk of having a stroke is bigger, especially at age 80 the peak is huge. Let's look at our target variable now.</p>

```python
df['stroke'].value_counts().plot.pie(title="Had a stroke", autopct='%1.1f%%')
```

<p style="text-align:justify;margin-right:100px;">Less than 5% of the patients in the dataset had a stroke. That means our dataset is really imbalanced. When we will try to predict something, the machine will almost always predict 'no', since we have much more of those in our datasets. Our dataset is really imbalanced now and we should make it more balanced. We will downsample our data, so we will remove a lot of rows of patients who did not have a stroke.</p>

```python
from sklearn.utils import resample

df_majority = df[df['stroke']==0]
df_minority = df[df['stroke']==1]
 
# Downsample majority class
df_majority_downsampled = resample(df_majority, 
                                 replace=False,    # sample without replacement
                                 n_samples=249,    # to match minority class
                                 random_state=0)   # to keep the same values everytime we run the notebook
                                 
df_downsampled = pd.concat([df_majority_downsampled, df_minority])
 
df_downsampled.stroke.value_counts()
```

<h2>Logistic Regression</h2><p style="text-align:justify;margin-right:100px;">We now continue with logistic regression as this algorithm is good for predicting yes or no outcomes. We start by selecting features which might be good.</p>

```python
# X = df_downsampled[['age', 'hypertension', 'heart_disease', 'avg_glucose_level', 'bmi']]
# X = df_downsampled[['hypertension', 'heart_disease', 'avg_glucose_level', 'bmi']]
# X = df_downsampled[['age', 'hypertension', 'heart_disease', 'avg_glucose_level']]
# X = df_downsampled[['hypertension', 'heart_disease', 'avg_glucose_level']]
# X = df_downsampled[['hypertension', 'heart_disease']]
# X = df_downsampled[['hypertension']]
# X = df_downsampled[['heart_disease']]
# X = df_downsampled[['avg_glucose_level']]
X = df_downsampled[['age', 'heart_disease', 'hypertension']]

y = df_downsampled["stroke"]
```

<p style="text-align:justify;margin-right:100px;">The selected features gave the best results. Before we actually apply Logistic Regression, we want to know what parameters will give us the best results. To do this, we perform grid search.</p>

```python
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)
```

```python
pipelineLR = make_pipeline(StandardScaler(), LogisticRegression(random_state=1, penalty='l2', solver='lbfgs'))

# Create the parameter grid
param_grid_lr = [{
    'logisticregression__C': [0.001, 0.005, 0.008, 0.01],
}]

# Create an instance of GridSearch Cross-validation estimator
gsLR = GridSearchCV(estimator=pipelineLR,
                     param_grid = param_grid_lr,
                     scoring='precision',
                     cv=10,
                     refit=True,
                     n_jobs=1)

# Train the LogisticRegression Classifier
gsLR = gsLR.fit(X_train, y_train)

# Print the training score of the best model
print(gsLR.best_score_)

# Print the model parameters of the best model
print(gsLR.best_params_)

# Print the test score of the best model
clfLR = gsLR.best_estimator_
print('Test accuracy: %.3f' % clfLR.score(X_test, y_test))
```

<p style="text-align:justify;margin-right:100px;">According to the grid search, the best model needs a hyperparameter (C) of 0.001 (by default our hyperparameter is 1.00). We applied precision for scoring, since this is what we want to get the best results on. We also apply solver='liblinear', which is good to use for small datasets.</p> <a href="https://ibmsoe.github.io/snap-ml-doc/sklogregdoc.html">https://ibmsoe.github.io/snap-ml-doc/sklogregdoc.html</a>

```python
model = LogisticRegression(solver='liblinear', random_state=0, C=0.005)
```

```python
model.fit(X, y)
```

```python
# probability estimates
model.predict_proba(X)
```

<p style="text-align:justify;margin-right:100px;">Now that we chose our hyperparameter, we can see that the model is guessing a lot, as many of them are close to 0.5. (Left is no, right is yes) It is mostly guessing 1 (stroke), which is better than predicting no stroke, when someone did have a stroke. Everything that was close to 1 when making the prediction, will be put to 1. But everything that is 0.5 or lower, will be predicted as no. If the result is around 0.55 or 0.6, it tried really hard to changed the value to no, but this did not work, which means if was a clear yes.</p>

```python
model.predict(X)
```

<p style="text-align:justify;margin-right:100px;">Above we can see what actually was predicted to be the outcome. We see that most of them are 1 (stroke).</p>

```python
model.score(X, y)
```

Because we chose a hyperparameter and tried to get the best results possible, either recall or precision will get better results. Since one of them gets better, the other one will get worse. In our case, it got much worse, since our accuracy decreased a lot.

```python
confusion_matrix(y, model.predict(X))
```

We will visualize the confusion matrix to get a better understanding of what was predicted.

```python
cm = confusion_matrix(y, model.predict(X))

fig, ax = plt.subplots(figsize=(5, 5))
ax.imshow(cm, cmap='tab20')
ax.grid(False)
ax.xaxis.set(ticks=(0, 1), ticklabels=('Predicted 0s', 'Predicted 1s'))
ax.yaxis.set(ticks=(0, 1), ticklabels=('Actual 0s', 'Actual 1s'))
ax.set_ylim(1.5, -0.5)
for i in range(2):
    for j in range(2):
        ax.text(j, i, cm[i, j], ha='center', va='center', color='black', fontsize='large')
plt.show()
```

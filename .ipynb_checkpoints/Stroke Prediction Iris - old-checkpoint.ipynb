<div id=domain-understanding><h2>Domain Understanding</h2></div>
<p style="text-align: justify;">To be able to predict whether someone will have a stroke, we first need to do some research about stroke in general. </p><br>

<b> What is a stroke?</b>
<p style="text-align: justify; margin-top:0;">A stroke (or brain attack) happens when either something blocks the blood supply to the brain or when a blood vessel in the brain bursts. </p><br>

1. <u>Ischemic stroke:</u>
<p style="text-align: justify; margin-top:0;">This happens when blood clots or other particles block the blood vessels to the brain. This type of stroke is the most common one.</p><br>

2. <u>Hemorrhagic stroke:</u>
<p style="text-align: justify; margin-top:0;">This stroke happens when an artery in the brain leaks blood or breaks open.</p><br>

3. <u>Transient ischemic stroke (TIA):</u>
<p style="text-align: justify; margin-top:0;margin-right: 100px;">Also called a 'mini-stroke'. When you have a TIA, it is basically a warning sign of a future stroke. The symptoms of a major stroke are similar to the symtoms of a TIA. This type of stroke is often caused by blood clots (like for ischemic stroke). Even though this type of stroke only occurs for a few minutes, it is very important to treat it, as more than third of people who had a TIA got a major stroke within 1 year. For 10% to 15% this was even within 3 months.</p>

<p style="text-align:justify;margin-right:100px;margin-top:20px;">
<b>Symptoms</b><br>
The most common symptoms are:
<ul>
    <li>Sudden numbness or weakness in the face, arm, or leg, especially on one side of the body.</li>
    <li>Sudden confusion, trouble speaking, or difficulty understanding speech.</li>
    <li>Sudden trouble seeing in one or both eyes.</li>
    <li>Sudden trouble walking, dizziness, loss of balance, or lack of coordination.</li>
    <li>Sudden severe headache with no known cause.</li>
</ul>
</p>

<p style="text-align:justify;margin-right:100px;margin-top:20px;">
<b>What could increase the chance of having a stroke?</b><br>
There are some things that can increase the chance of having a stroke. 
To give a few examples:
<ul>
    <li style="text-align:justify;margin-right:100px;">Hypertension: high blood pressure damages arteries throughout the body, creating conditions in which they can burst or clog more easily. Weakened or blocked arteries in the brain put you at a much higher risk for stroke, which is why managing high blood pressure is critical to reducing your stroke risk.</li>
    <li style="text-align:justify;margin-right:100px;">Heart disease: common heart disorders can increase your risk for stroke. For example, coronary artery disease increases your risk for stroke, because plaque builds up in the arteries and blocks the flow of oxygen-rich blood to the brain. Other heart conditions, such as heart valve defects, irregular heartbeat (including atrial fibrillation), and enlarged heart chambers, can cause blood clots that may break loose and cause a stroke.</li>
    <li style="text-align:justify;margin-right:100px;">Smoking: damages blood vessels, which can cause a stroke. About 1 in 7 men smoke. Men are also more likely to be smokers than women are.</li>
    <li style="text-align:justify;margin-right:100px;">Overweight and obesity: having overweight or obesity increases stroke risk. About 3 in 4 men in the United States have overweight or obesity.</li>
    <li style="text-align:justify;margin-right:100px;">Diabetes: increases stroke risk because it can harm blood vessels in the brain. About 1 in 7 men have diabetes.</li>
    <li style="text-align:justify;margin-right:100px;">Age: you are more likely to have a stroke if you are older. It is even so that the chance of having a stroke doubles every 10 years after age 55. However, nowadays it is so that one in seven strokes occur in adolescents and young adults ages 15-49, this could be because more young people have obesity, high blood pressure and diabetes.</li>
    <li style="text-align:justify;margin-right:100px;">Sex: women are more likely to have a stroke, this is because pregnancy and birth control pills have an extra risk.</li>
</ul>
</p>
<p style="text-align:justify;margin-right:100px;margin-top:20px;">
<b>What to do when you notice any of the symptoms?</b><br>
If a person might have a stroke, think <b>F.A.S.T.</b>:
<ul>
    <li><u><b>F</b>ace:</u> ask the person to smile to check whether one side of the face is drooping</li>
    <li><u><b>A</b>rms:</u> ask the person to raise both arms, and see if one arm drifts downwards.</li>
    <li><u><b>S</b>peech:</u> ask the person to repeat a short sentence to find out if their words sounds slurred.</li>
    <li><u><b>T</b>ime:</u> if you notice any of these signs, call the emergency assistance immediately.<br><br></li>
</ul>

<b>Treatment</b><br>
<p style="text-align:justify;margin-right: 100px;">It is crucial to call the emergency assistence for help immediately, as every minute counts. The faster a stroke is treated, the less damage is done to the brain. Some treatments can only be used if the person reaches the hospital in the first three hours after symptoms appear.</p>

<p style="text-align:justify;margin-right:100px;">Since the dataset is really imbalanced, we should make it more balanced, since in the end the machine will most likely 
look at most 0 (no). I tried to use down-sample on our minority class, since there are way too many values in the majority 
class. This is actually not that good, since we are losing a lot of data, but for the time being we do this. </p>

```python
from sklearn.utils import resample

df_majority = df[df.stroke==0]
df_minority = df[df.stroke==1]
 
# Downsample majority class
df_majority_downsampled = resample(df_majority, 
                                 replace=False,    # sample without replacement
                                 n_samples=249,    # to match minority class
                                 random_state=0)   # to keep the same values everytime we run the notebook
                                 
# Combine minority class with downsampled majority class
df_downsampled = pd.concat([df_majority_downsampled, df_minority])
 
# Display new class counts
df_downsampled.stroke.value_counts()
```

<div id=modelling><h2>Modelling</h2></div>
<p style="text-align:justify;margin-right: 50px;">
    For iteration 0 we will only use Nearest Neighbors since it is an easy algorithm to start with. In further iterations we will look at different iterations.
</p>

```python
X = df[['gender', 'age', 'hypertension', 'heart_disease', 
        'ever_married', 'work_type', 'Residence_type', 
        'avg_glucose_level', 'bmi', 'smoking_status']]
y = df["stroke"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print("There are in total", len(X), "observations.")
print("The train set now has", len(X_train), "observations.")
print("The test set now has", len(X_test), "observations.")
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_train, y_train)
pred = model.predict(X_test)

from sklearn.metrics import accuracy_score
acc = accuracy_score(pred, y_test)
print("Accuracy:", acc)
```

<p style="text-align: justify;margin-right:100px;">The accuracy is almost 100, which is really high. However, in this case it is actually not good because of the imbalanced dataset. Because the dataset is so imbalanced, the machine will look at most <b>no</b>. This causes the accurary to be high since the machine is only looking at no. Now we will look at Nearest Neighbors with the balanced dataset and see if we do actually get a lower accuracy.</p>

```python
X = df_downsampled[['gender', 'age', 'hypertension', 'heart_disease', 
                    'ever_married', 'work_type', 'Residence_type', 
                    'avg_glucose_level', 'bmi', 'smoking_status']]
y = df_downsampled["stroke"]
```

```python
# from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print("There are in total", len(X), "observations.")
print("The train set now has", len(X_train), "observations.")
print("The test set now has", len(X_test), "observations.")
```

```python
#from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_train, y_train)
pred = model.predict(X_test)

#from sklearn.metrics import accuracy_score
acc = accuracy_score(pred, y_test)
print("Accuracy:", acc)
```

<p style="text-align:justify;margin-right:100px;">The accuracy decreased to 0.7. That is, even though it is lower, much better, than an accuracy of 0.95 with the imbalanced dataset. Now the machine actually looks at evenly divided data of the target variable. </p>
